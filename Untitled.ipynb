```python
import torch
import torch.nn as nn
import torchvision
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_batch_train = 5000
num_batch_val = 1000

train_batch = None
for i in range(2, 6):
    with open('cifar-10-batches-py/data_batch_{0}'.format(i), 'rb') as f:
        batch = pickle.load(f, encoding='bytes')
        if train_batch is not None:
            train_batch[b'data'] = np.concatenate((train_batch[b'data'], batch[b'data']))
            train_batch[b'labels'] += batch[b'labels']
        else:
            train_batch = batch

with open('cifar-10-batches-py/test_batch', 'rb') as f:
    test_batch = pickle.load(f, encoding='bytes')

to_tensor = torchvision.transforms.ToTensor()
train_data = to_tensor(train_batch[b'data'][0:num_batch_train]).squeeze()
train_labels = torch.LongTensor(train_batch[b'labels'][0:num_batch_train])
test_data = to_tensor(train_batch[b'data'][num_batch_train:num_batch_train+num_batch_val]).squeeze()
test_labels = torch.LongTensor(train_batch[b'labels'][num_batch_train:num_batch_train+num_batch_val])

del train_batch
del test_batch
```

```python
def train(net, lr=1e-4, lr_decay=1, num_iter=100, batch_size=250):
    optim = torch.optim.SGD(net.parameters(), momentum=0, lr=lr)
    lr_scheduler = torch.optim.lr_scheduler.StepLR(optim, step_size=1, gamma=lr_decay)
    loss_fn = torch.nn.CrossEntropyLoss()
    
    train_loss_history = []
    val_loss_history = []
    
    batch_sel = 0
    for i in range(num_iter):
        with torch.no_grad():
            if i % (num_iter // 100) == 0:
                net.eval()
                pred = net(test_data)
                loss = loss_fn(pred, test_labels)
                val_loss_history.append(loss)
        
        net.train()
        pred = net(train_data[batch_sel : batch_sel + batch_size])
        loss = loss_fn(pred, train_labels[batch_sel : batch_sel + batch_size])
        
        if i % (num_iter // 100) == 0:
            train_loss_history.append(loss)
        
        loss.backward()
        optim.step()
        lr_scheduler.step()
        
        batch_sel += batch_size
        if batch_sel + batch_size > train_data.shape[0]:
            batch_sel = 0
    
    plt.plot(train_loss_history, 'b', val_loss_history, 'r--')
    
        
        
```

```python
net = nn.Sequential(
    nn.Linear(3072, 200),
    nn.ReLU(True),
    nn.Linear(200, 100),
    nn.ReLU(True),
    nn.Linear(100, 10)
)
train(net, lr=5e-4, lr_decay=0.9999, num_iter=2000)
```
